```python
import pickle as pk
import matplotlib.pyplot as plt
```

```python
with open('losses_pickled/fake_loss_disc_12e','rb') as f:
    fkd = pk.load(f)
```

```python
with open('losses_pickled/real_loss_disc_12e','rb') as f:
    rkd = pk.load(f)
```

```python
with open('losses_pickled/real_loss_gen_12e','rb') as f:
    rkg = pk.load(f)
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
erkg = []
for i in range(12):
    erkg.append(np.mean(rkg[12*i:12*(i+1)]))

fig = plt.figure()
plt.plot(erkg, label="Loss")
fig.suptitle('Generator Loss (Lambda = 0.65)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
fig.savefig('GLoss')
```

```python
efkd = []
for i in range(12):
    efkd.append(np.mean(fkd[12*i:12*(i+1)]))

fig = plt.figure()
plt.plot(efkd, label="Loss")
fig.suptitle('Discriminator Loss (On fake inputs)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
fig.savefig('Dlossfake')
```

```python
erkd = []
for i in range(12):
    erkd.append(np.mean(rkd[12*i:12*(i+1)]))

fig = plt.figure()
plt.plot(erkd, label="Loss")
fig.suptitle('Discriminator Loss (On real inputs)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
fig.savefig('Dlossreal')
```
